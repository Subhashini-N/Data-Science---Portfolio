```python
# Import required libraries

import pandas as pd
from matplotlib import pyplot as plt 
```

```python
# Load movies data into dataframe

movies_df = pd.read_csv("/Users/manikantanvenkataraman/Downloads/ml-latest-small/movies.csv")
movies_df.head()
```

```python
#Load ratings data into a dataframe

ratings_df = pd.read_csv("/Users/manikantanvenkataraman/Downloads/ml-latest-small/ratings.csv")
ratings_df.head()
```

```python
# Load tags data into a dataframe

tags_df = pd.read_csv("/Users/manikantanvenkataraman/Downloads/ml-latest-small/tags.csv")
tags_df.head()
```

```python
# Load links data into a dataframe

links_df = pd.read_csv("/Users/manikantanvenkataraman/Downloads/ml-latest-small/tags.csv")
links_df.head()
```

```python
# Merge movies dataframe and ratings dataframe

movie_rating_1 =pd.merge(movies_df,ratings_df, on='movieId')
movie_rating_1.head()
```

```python
# Merge dataframe from above and links dataframe

movie_rating_2=pd.merge(movie_rating_1,links_df, on=["movieId", "userId"])
movie_rating_2.head()
```

```python
# Merge dataframe from above and tags dataframe to create a final dataframe

movie_rating =pd.merge(movie_rating_2,tags_df, on=["movieId", "userId"])
movie_rating.head()
```

```python
# Obtain average ratings by title

Average_ratings = pd.DataFrame(movie_rating.groupby('title')['rating'].mean())
Average_ratings.head()
```

```python
# Obtain total number of ratings for a title

Average_ratings['Total Ratings'] = pd.DataFrame(movie_rating.groupby('title')['rating'].count())
Average_ratings.head()
```

```python
# Create a pivot table to list the movies and rating for every user

movie_user = movie_rating.pivot_table(index='userId',columns='title',values='rating')
movie_user.head()
```

```python
# Calculate Correlation

correlations = movie_user.corrwith(movie_user['2001: A Space Odyssey (1968)'])
correlations.head(10)
```

```python
# Obtain recommendation movies for the chosen movie

recommendation = pd.DataFrame(correlations,columns=['Correlation'])
recommendation.dropna(inplace=True)
recommendation = recommendation.join(Average_ratings['Total Ratings'])
recommendation.head(10)
```

```python
# Define function to calculate recommendation
def movie_recommender(title):
    correlations = movie_user.corrwith(title)
    recommendation = pd.DataFrame(correlations,columns=['Correlation'])
    recommendation.dropna(inplace=True)
    recommendation = recommendation.join(Average_ratings['Total Ratings'])
    return recommendation.title
    
```

```python
# Obtain user input to determine recommendation
def main():
    movie_input = input('Enter a moviename:')
    f= movie_recommender(movie_input)
    print(f)
    
main()
```
